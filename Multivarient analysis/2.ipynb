# Box-plot (Numerical-Categorical)

```python
# Compare how many male and female survived 
sns.boxplot(x=titanic['Sex'],y=titanic['Age'])
```

```python
# what age of male/females did survive
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

# Dist-Plot (Numerical-Categorical)

```python
# Probability of passengers surviving on the basis of Age
sns.distplot(x=titanic[titanic['Survived']==0]['Age'],hist=None)
sns.distplot(x=titanic[titanic['Survived']==1]['Age'],hist=None) 
```

```python
sns.get_dataset_names()
```

```python
# Probablity of male and female survived according to there age
sns.distplot(x=titanic[titanic['Sex']=='male']['Survived'],hist=None)
sns.distplot(x=titanic[titanic['Sex']=='female']['Survived'],hist=None) 
```

# HeatMap (Categorical-Categorical)

```python
# find out how many people survived and how many people did not survive in each class
# using crosstab functionality
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived'],))
```

```python
# How much percent of people survived in Pclass
(titanic.groupby('Pclass')['Survived'].mean()*100).plot(kind='pie',autopct='%0.2f')
```

```python
# How much percent of people survived from male and female
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='pie',autopct='%0.02f')
```

```python
# How much percent of people survived in Who embarked from S. C, Q
(titanic.groupby('Embarked')['Survived'].mean()*100).plot(kind='pie',autopct='%0.2f')
```

# Cluster-Map (Category-Category)


```python
# how may people survived who were accompanied with there siblings and spouse
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

```python
# how many people survived who accompanied with there parent or child
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

# Line-Plot (Numrical-Numerical)

```python
# line plot is used when you have a quantity in the x-axis time based like (time, month, year etc)
flights.head(10)
```

```python
# Selecting only those columns which contains only numerical
numerical_columns=flights.select_dtypes(include=['number'])
# perform groupby and sum operations only on numerical column
result=numerical_columns.groupby(flights['year']).sum().reset_index()
```

```python
result
```

```python
sns.lineplot(data=flights, x=flights['year'],y=flights['passengers'])
```

```python
# using heatmap 
flights.pivot_table(values='passengers',index='month',columns='year')
```
